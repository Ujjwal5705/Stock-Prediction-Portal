```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime 
```

# 1.Data Collection

```python
now = datetime.now()

ticker = 'AAPL'
end = now
start = datetime(now.year-10, now.month, now.day)
df = yf.download(ticker, start, end, auto_adjust=False)
df
```
