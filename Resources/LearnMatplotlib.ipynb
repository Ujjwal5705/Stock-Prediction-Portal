### Subplots & Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 15, 25, 30, 40]
predicted_price = [8, 16, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.title('plot with legends');
plt.xlabel('Days');
plt.ylabel('Price');
plt.legend();
```

#### Subplots

```python
# Subplot for Original Price
plt.subplot(3, 1, 1);
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.title('Subplot for Original Price');
plt.xlabel('Days');
plt.ylabel('Price');
plt.legend();


# Subplot for Predicted Price
plt.subplot(3, 1, 2);
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.title('Subplot for Predicted Price');
plt.xlabel('Days');
plt.ylabel('Price');
plt.legend();


# Subplot for difference in Price
plt.subplot(3, 1, 3);
diff_arr = [o-p for o,p in zip(original_price, predicted_price)]
plt.bar(days, diff_arr, color='purple', label='Price Difference');
plt.title('Subplot for difference in original and predicted price');
plt.xlabel('Days');
plt.ylabel('Price Difference');
plt.legend();


plt.tight_layout()
```

### Saving the figure

```python
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.plot(days, predicted_price, 'r-o', label='Predicted Price');
plt.title('plot with legends');
plt.xlabel('Days');
plt.ylabel('Price');
plt.legend();
plt.savefig('example_prediction_3.png', dpi=600)
```
